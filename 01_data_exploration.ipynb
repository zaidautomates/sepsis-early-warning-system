```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries loaded successfully")

# ── Load all .psv files from setA and setB ──────────────────────
def load_psv_files(folder_path):
    all_dfs = []
    files = [f for f in os.listdir(folder_path) if f.endswith('.psv')]
    print(f"Found {len(files)} patient files in {folder_path}")
    
    for i, filename in enumerate(files):
        filepath = os.path.join(folder_path, filename)
        df = pd.read_csv(filepath, sep='|')
        
        # Add patient ID from filename
        patient_id = filename.replace('.psv', '')
        df['PatientID'] = patient_id
        
        # Add hour index
        df['Hour'] = range(len(df))
        
        all_dfs.append(df)
    
    combined = pd.concat(all_dfs, ignore_index=True)
    return combined

# Update these paths to match where you extracted the files
setA_path = '../data/setA'
setB_path = '../data/setB'

df_A = load_psv_files(setA_path)
df_B = load_psv_files(setB_path)

# Combine both sets
df = pd.concat([df_A, df_B], ignore_index=True)
print(f"\nTotal rows loaded: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
print(f"Total unique patients: {df['PatientID'].nunique():,}")
```

```python
# ── Basic data overview ──────────────────────────────────────────
print("=== DATASET OVERVIEW ===\n")
print(df.head(3))
print("\n=== COLUMN NAMES ===")
print(df.columns.tolist())
print("\n=== DATA TYPES ===")
print(df.dtypes)
print("\n=== MISSING VALUES (%) ===")
missing = (df.isnull().sum() / len(df) * 100).round(2)
print(missing[missing > 0].sort_values(ascending=False))
```
